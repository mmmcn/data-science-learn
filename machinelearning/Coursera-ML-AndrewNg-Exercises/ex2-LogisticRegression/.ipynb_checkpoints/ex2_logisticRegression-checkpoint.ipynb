**实现sigmiod函数功能，以提供后续调用**

```python
def sigmoid(z):
    return 1 / (1 + np.exp(-z))

nums = np.arange(-10, 10, step=1)
fig, ax = plt.subplots(figsize=(12,8))
ax.plot(nums, sigmoid(nums), 'r')
plt.show()
```

**代价函数**    
$$J\left( \theta  \right)=\frac{1}{m}\sum\limits_{i=1}^{m}{[-{{y}^{(i)}}\log \left( {{h}_{\theta }}\left( {{x}^{(i)}} \right) \right)-\left( 1-{{y}^{(i)}} \right)\log \left( 1-{{h}_{\theta }}\left( {{x}^{(i)}} \right) \right)]}$$

```python
def cost(theta, x, y):
    theta = np.matrix(theta)
    x = np.matrix(x)
    y = np.matrix(y)
    first_term = np.multiply(-y, np.log(sigmoid(x * theta.T)))
    second_term = np.multiply((1 - y), np.log(1 - sigmoid(x * theta.T)))
    
    return np.sum(first_term - second_term) / len(x)          # x为矩阵形式，len(x)返回矩阵的行数
```

```python
data.insert(0, 'ones', 1)
data.head()
```

```python
data.shape
```

```python
#从数据集中截取出x和y
cols = data.shape[1]
x = data.iloc[:,0:cols-1]
y = data.iloc[:,cols-1:cols]
```

```python
x = np.array(x.values)
y = np.array(y.values)
theta = np.zeros(3)
```

```python
theta.shape, x.shape, y.shape
```

```python
#初始化参数theta时的代价
cost(theta, x, y)
```

# 执行Gradient descent找到最优化参数

```python
def gradient(theta, x, y):
    theta = np.matrix(theta)
    x = np.matrix(x)
    y = np.matrix(y)
    
    parameters = int(theta.ravel().shape[1])
    grad = np.zeros(parameters)
    
    error = sigmoid(x * theta.T) - y
    
    for i in range(parameters):
        term = np.multiply(error, x[:,i])
        grad[i] = np.sum(term) / len(x)
        
    return grad
```

```python
gradient(theta, x, y)         #这里实际输出的是对应参数下梯度步长
```

```python
import scipy.optimize as opt
result = opt.fmin_tnc(func=cost, x0=theta, fprime=gradient, args=(x,y))
result
```

```python
cost(result[0], x, y)
```

```python
def predict(theta, x):
    probability = sigmoid(x * theta.T)
    return [1 if x >= 0.5 else 0 for x in probability]
```

```python
theta_min = np.matrix(result[0])
predictions = predict(theta_min, x)
accuracy = np.sum((np.array(predictions) - y.ravel()) == 0)    #此处需注意y的shape应与predictions的shape保持一致
print('accuracy = {0}%'.format(accuracy))
```

```python
degree = 6
x1 = data2['Test1']
x2 = data2['Test2']

data2.insert(3, 'Ones', 1)

for i in range(1, degree + 1):
    for j in range(0, i + 1):
        data2['F' + str(i - j) +str(j)] = np.power(x1, i -j) * np.power(x2, j)
```

```python
data2.drop('Test1', axis=1, inplace=True)
data2.drop('Test2', axis=1, inplace=True)

data2.head()
```

# 正则化代价参数
$$J\left( \theta  \right)=\frac{1}{m}\sum\limits_{i=1}^{m}{[-{{y}^{(i)}}\log \left( {{h}_{\theta }}\left( {{x}^{(i)}} \right) \right)-\left( 1-{{y}^{(i)}} \right)\log \left( 1-{{h}_{\theta }}\left( {{x}^{(i)}} \right) \right)]}+\frac{\lambda }{2m}\sum\limits_{j=1}^{n}{\theta _{j}^{2}}$$

```python
def cost(theta, x, y, learningRate):               #参数中的learningRate即为代价函数中的λ
    theta = np.matrix(theta)
    x = np.matrix(x)
    y = np.matrix(y)
    first_term = np.multiply(-y, np.log(sigmoid(x * theta.T)))
    second_term = np.multiply((1 - y), np.log(1 - sigmoid(x * theta.T)))
    reg = (learningRate / (2 * len(x))) * np.sum(np.power(theta[:,1:theta.shape[1]], 2))
    
    return np.sum(first_term - second_term) / len(x) + reg
```

# 使用梯度下降法最小化代价函数
这里需要区别$\theta_0$以及$\theta_j \left( j=1,2,3,... \right)$的更新情况

```python
def gradReg(theta, x, y, learningRate):
    theta = np.matrix(theta)
    x = np.matrix(x)
    y = np.matrix(y)
    
    parameters = int(theta.ravel().shape[1])
    grad = np.zeros(parameters)
    
    error = sigmoid(x * theta.T) - y
    
    for i in range(parameters):
        term = np.multiply(error, x[:,i])
        
        if (i == 0):
            grad[i] = np.sum(term) / len(x)
        else:
            grad[i] = (np.sum(term) / len(x)) + ((learningRate / len(x)) * theta[:,i])
    
    return grad
```

```python
# set X and y (remember from above that we moved the label to column 0)
cols = data2.shape[1]
x2 = data2.iloc[:,1:cols]
y2 = data2.iloc[:,0:1]

# convert to numpy arrays and initalize the parameter array theta
x2 = np.array(x2.values)
y2 = np.array(y2.values)
```

```python
theta2 = np.zeros(28)
learningRate = 1
```

```python
cost(theta2, x2, y2, learningRate)
```

```python
gradReg(theta2, x2, y2, learningRate)
```

```python
result2 = opt.fmin_tnc(func=cost, x0=theta2, fprime=gradReg, args=(x2,y2,learningRate))
result2
```

# 得到的预测准确度还挺高的

```python
theta_min = np.matrix(result2[0])
predictions = predict(theta_min, x2)
accuracy = np.sum((np.array(predictions) - y2.ravel()) == 0)    
print('accuracy = {0}%'.format(accuracy))
```
