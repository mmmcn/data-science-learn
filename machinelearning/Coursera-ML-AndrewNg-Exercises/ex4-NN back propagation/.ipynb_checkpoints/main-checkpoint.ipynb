```python
def load_data(path, transpose=True):
    data = sio.loadmat(path)
    X = data.get('X')   # (5000,400)
    y = data.get('y')   # (5000,1)
    y = y.reshape(y.shape[0])  # (5000,)
    
    if transpose:
        X = np.array([im.reshape((20, 20)).T for im in X])  # (5000, 20, 20)
        
        X = np.array([im.reshape(400) for im in X])    # (5000, 400)
    return X, y
```

```python
def sigmoid(z):    # 激励函数
    return 1 / (1 + np.exp(-z))
```

# 前向传播函数

```python
def forward_propagate(X, thetal, theta2):
    '''
    theta1: 存储第一层权重参数
    theta2: 存储第二次权重参数
    '''
    m = X.shape[0]
    
    a1 = np.insert(X, 0, values=np.ones(m), axis=1)
    z2 = a1 * theta1.T     # (5000,401) * (401,25) = (5000,25)
    a2 = np.insert(sigmoid(z2), 0, values=np.ones(m), axis=1)
    z3 = a2 * theta2.T
    h = sigmoid(z3)
    
    return a1, z2, a2, z3, h
```

# 代价函数

```python
def cost(params, input_size, hidden_size, num_labels, X, y, learning_rate):
    '''
    传入的params存储了每一层中连接的权重
    使用时需按照每层连接参数个数取出
    '''
    m = X.shape[0]
    X = np.matrix(X)
    y = np.matrix(y)
    
    theta1 = np.matrix(np.reshape(params[:hidden_size * (input_size + 1)], (hidden_size, (input_size + 1))))
    theta2 = np.matrix(np.reshape(params[hidden_size * (input_size + 1):], (num_labels, (hidden_size + 1))))

    # 进行前向传播
    a1, z2, a2, z3, h = forward_propagate(X, theta1, theta2)
    
    # compute the cost 
    J = 0
    for i in range(m):     # 所有训练数据
        first = np.multiply(-y[i,:], np.log(h[i,:]))
        second = np.multiply((1 - y[i,:]), np.log(1 - h[i,:]))
        J += np.sum(first - second)   # 相当于公式中的内层求和
    J = J / m
    
    return J
```

```python
# 初始参数设置
input_size = 400
hidden_size = 25
num_labels = 10
learning_rate = 1

# 随机初始化网络参数，注意参数不能相同
params = (np.random.random(size=hidden_size * (input_size + 1) + num_labels * (hidden_size + 1)) - 0.5) * 0.25

m = X.shape[0]
X = np.matrix(X)
y = np.matrix(y)

# 提取参数数组中元素到theta1和theta2中
theta1 = np.matrix(np.reshape(params[:hidden_size * (input_size + 1)], (hidden_size, (input_size + 1))))
theta2 = np.matrix(np.reshape(params[hidden_size * (input_size + 1):], (num_labels, (hidden_size + 1))))

theta1.shape, theta2.shape
```

# 加入正则化的代价函数

```python
def costWithReg(params, input_size, hidden_size, num_labels, X, y, learning_rate):
    m = X.shape[0]
    X = np.matrix(X)
    y = np.matrix(y)
    
    theta1 = np.matrix(np.reshape(params[:hidden_size * (input_size + 1)], (hidden_size, (input_size + 1))))
    theta2 = np.matrix(np.reshape(params[hidden_size * (input_size + 1):], (num_labels, (hidden_size + 1))))

    # 进行前向传播
    a1, z2, a2, z3, h = forward_propagate(X, theta1, theta2)
    
    # compute the cost   ,by loop
    J = 0
    for i in range(m):
        first = np.multiply(-y[i,:], np.log(h[i,:]))
        second = np.multiply((1 - y[i,:]), np.log(1 - h[i,:]))
        J += np.sum(first - second)
    J = J / m
    
    # 加入正则化项,以上部分与未正则化之前完全一致
    
    J += (float(learning_rate) / (2 * m) / (np.sum(np.power(theta1[:,1:], 2)) + np.sum(np.power(theta2[:,1:], 2))))
    
    return J
```

```python
costWithReg(params, input_size, hidden_size, num_labels, X, y_onehot, learning_rate)
```

# 反向传播算法

**To help you get started with this part of the exercise, you will first implement
the sigmoid gradient function. The gradient for the sigmoid function can be
computed as**  
$$g^`(z) = \frac{d}{dz}{g\left( z \right)}= {g\left( z \right)}{\left( {1 - {g\left( z \right)}}\right )}$$  
**where**  
$$ sigmoid(z) = g(z)={\frac{1}{1+e^{-z}}}$$


```python
def sigmoid_gradient(z):
    return np.multiply(sigmoid(z), (1 - sigmoid(z)))
```

```python
def back_propagate(params, input_size, hidden_size, num_labels, X, y, learning_rate):
    m = X.shape[0]   #数据个数
    X = np.matrix(X)
    y = np.matrix(y)
    
    # 之前验证了随机初始的params存储形式为一维
    # 从params中取出每一层的参数，以矩阵形式存储
    theta1 = np.matrix(np.reshape(params[:hidden_size * (input_size + 1)], (hidden_size, (input_size + 1))))
    theta2 = np.matrix(np.reshape(params[hidden_size * (input_size + 1):], (num_labels, (hidden_size + 1))))
    
    # 前向传播  ((5000, 401), (5000, 25), (5000, 26), (5000, 10), (5000, 10))
    a1, z2, a2, z3, h = forward_propagate(X , theta1, theta2)
    
    # J为代价函数值，delta存储“error term”
    J = 0
    delta1 = np.zeros(theta1.shape)  # (25, 401)
    delta2 = np.zeros(theta2.shape)  # (10, 26)
    
    # 计算代价函数值，这里未调用之前写好的，之后优化好参数啥的直接调用
    for i in range(m):
        first = np.multiply(-y[i,:], np.log(h[i,:]))
        second = np.multiply((1 - y[i,:]), np.log(1 - h[i,:]))
        J += np.sum(first - second)
        
    J = J / m
    
    # 加入正则化项
    
    J += (float(learning_rate) / (2 * m) / (np.sum(np.power(theta1[:,1:], 2)) + np.sum(np.power(theta2[:,1:], 2))))
    
    # 以下部分为BP算法
    for k in range(m):
        a1_k = a1[k, :]  # (1,401)
        z2_k = z2[k, :]  # (1,25)
        a2_k = a2[k, :]  # (1,26)
        # z3_k = z3[k, :]  # (1,10)
        h_k = h[k, :]  # (1,10)
        y_k = y[k, :]  # (1,10)
        
        d3_k = h_k - y_k  # (1,10)
        
        z2_k = np.insert(z2_k, 0, values=np.ones(1))   # (1, 26)
        d2_k = np.multiply((theta2.T * d3_k.T).T, sigmoid_gradient(z2_k))
        
        delta1 = delta1 + (d2_k[: ,1:]).T * a1_k
        delta2 = delta2 + d3_k.T * a2_k
    
    delta1 = delta1 / m
    delta2 = delta2 / m
    
    # print(np.ravel(delta1).shape, np.ravel(delta2).shape)
    
    grad = np.concatenate((np.ravel(delta1), np.ravel(delta2)))
    
    return J, grad
        
```

将梯度计算加入正则化处理

```python
def back_propagate_reg(params, input_size, hidden_size, num_labels, X, y, learning_rate):
    m = X.shape[0]   #数据个数
    X = np.matrix(X)
    y = np.matrix(y)
    
    # 之前验证了随机初始的params存储形式为一维
    # 从params中取出每一层的参数，以矩阵形式存储
    theta1 = np.matrix(np.reshape(params[:hidden_size * (input_size + 1)], (hidden_size, (input_size + 1))))
    theta2 = np.matrix(np.reshape(params[hidden_size * (input_size + 1):], (num_labels, (hidden_size + 1))))
    
    # 前向传播  ((5000, 401), (5000, 25), (5000, 26), (5000, 10), (5000, 10))
    a1, z2, a2, z3, h = forward_propagate(X , theta1, theta2)
    
    # J为代价函数值，delta存储“error term”
    J = 0
    delta1 = np.zeros(theta1.shape)  # (25, 401)
    delta2 = np.zeros(theta2.shape)  # (10, 26)
    
    # 计算代价函数值，这里未调用之前写好的，之后优化好参数啥的直接调用
    for i in range(m):
        first = np.multiply(-y[i,:], np.log(h[i,:]))
        second = np.multiply((1 - y[i,:]), np.log(1 - h[i,:]))
        J += np.sum(first - second)
        
    J = J / m
    
    # 加入正则化项
    
    J += (float(learning_rate) / (2 * m) / (np.sum(np.power(theta1[:,1:], 2)) + np.sum(np.power(theta2[:,1:], 2))))
    
    # 以下部分为BP算法
    for k in range(m):
        a1_k = a1[k, :]  # (1,401)
        z2_k = z2[k, :]  # (1,25)
        a2_k = a2[k, :]  # (1,26)
        z3_k = z3[k, :]  # (1,10)
        h_k = h[k, :]  # (1,10)
        y_k = y[k, :]  # (1,10)
        
        d3_k = h_k - y_k  # (1,10)
        
        z2_k = np.insert(z2_k, 0, values=np.ones(1))   # (1, 26)
        d2_k = np.multiply((theta2.T * d3_k.T).T, sigmoid_gradient(z2_k))
        
        delta1 = delta1 + (d2_k[: ,1:]).T * a1_k
        delta2 = delta2 + d3_k.T * a2_k
    
    delta1 = delta1 / m
    delta2 = delta2 / m
    
    # print(np.ravel(delta1).shape, np.ravel(delta2).shape)
    
    delta1[:,1:] = delta1[:,1:] + (theta1[:,1:] * learning_rate) / m
    delta2[:,1:] = delta2[:,1:] + (theta2[:,1:] * learning_rate) / m
    
    grad = np.concatenate((np.ravel(delta1), np.ravel(delta2)))
    
    return J, grad
        
```

```python
J ,grad = back_propagate_reg(params, input_size, hidden_size, num_labels, X, y_onehot, learning_rate)
J, grad.shape
```

```python
fmin = opt.minimize(fun=back_propagate_reg,
                   x0=params,
                   args=(input_size, hidden_size, num_labels, X, y_onehot,learning_rate),
                   method='TNC',
                   jac=True,
                   options={'maxiter' : 250}
                   )
fmin
```

```python
X = np.matrix(X)
theta1 = np.matrix(np.reshape(fmin.x[:hidden_size * (input_size + 1)], (hidden_size, (input_size + 1))))
theta2 = np.matrix(np.reshape(fmin.x[hidden_size * (input_size + 1):], (num_labels, (hidden_size + 1))))

a1, z2, a2, z3, h = forward_propagate(X, theta1, theta2)
y_pred = np.array(np.argmax(h, axis=1) + 1)
y_pred
```

```python
correct = [1 if a == b else 0 for (a, b) in zip(y_pred, y)]
accuracy = (sum(map(int, correct)) / float(len(correct)))
print ('accuracy = {0}%'.format(accuracy * 100))
```
